```python
df_raw = spark.read \
.format("csv") \
.option("header","true") \
.option("inferSchema","false") \
.load("/Volumes/etl_catalog/bronze/raw_csv_files/mandat_uzbmb_uz_2025/")
```

```python
 df_raw.show()
```

```python
df_raw.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("etl_catalog.bronze.students_raw")
```

```python
df_raw = df_raw.toDF(*[
    c.replace(" ", "_")
     .replace("'", "")
     .replace("-", "_")
    for c in df_raw.columns
])
```

```python
df_raw.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("etl_catalog.bronze.students_raw")
```

```python
bronze_df = spark.table("etl_catalog.bronze.students_raw")
```

```python
print(bronze_df.columns)
bronze_df.show(5)
```

```python
from pyspark.sql.functions import when, col, regexp_replace

df_silver = bronze_df \
.withColumn(
    "status",
    when(col("TR_class") == "table-success", "Grant")
    .when(col("TR_class") == "table-warning", "Kontrakt")
    .otherwise("Yiqilgan")
) \
.withColumn(
    "Ball",
    regexp_replace(col("Ball"), ",", ".").cast("float")
)
```

```python
df_silver.show(5)
```

```python
df_silver.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("etl_catalog.silver.students_cleaned")
```

```python
%sql
SELECT
    status,
    COUNT(*) AS student_count,
    ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER (), 2) AS percent
FROM etl_catalog.silver.students_cleaned
GROUP BY status;
```

```python
%sql
SELECT 
    Oliy_talim_muassasasi AS university,
    ROUND(AVG(Ball), 2) AS avg_grant_score
FROM etl_catalog.silver.students_cleaned
WHERE status = 'Grant'
GROUP BY Oliy_talim_muassasasi
ORDER BY avg_grant_score DESC
LIMIT 1;
```

```python
%sql
SELECT *
FROM etl_catalog.silver.students_cleaned
WHERE Yonalish = 'Iqtisodiyot'
  AND status = 'Yiqilgan'
ORDER BY Ball DESC
LIMIT 1;
```

```python
%sql
SELECT 
    Talim_tili AS language,
    COUNT(*) AS grant_count,
    ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER (), 2) AS grant_percent
FROM etl_catalog.silver.students_cleaned
WHERE status = 'Grant'
GROUP BY Talim_tili;
```

```python
status_gold_df = spark.sql("""
SELECT
    status,
    COUNT(*) AS student_count,
    ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER (), 2) AS percent
FROM etl_catalog.silver.students_cleaned
GROUP BY status
""")

status_gold_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("etl_catalog.gold.status_distribution")
```

```python
university_gold_df = spark.sql("""
SELECT 
    Oliy_talim_muassasasi AS university,
    ROUND(AVG(Ball), 2) AS avg_grant_score
FROM etl_catalog.silver.students_cleaned
WHERE status = 'Grant'
GROUP BY Oliy_talim_muassasasi
ORDER BY avg_grant_score DESC
""")

university_gold_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("etl_catalog.gold.university_ranking")
```

```python
fail_iqtisodiyot_gold_df = spark.sql("""
SELECT *
FROM etl_catalog.silver.students_cleaned
WHERE Yonalish = 'Iqtisodiyot'
  AND status = 'Yiqilgan'
ORDER BY Ball DESC
LIMIT 1
""")

fail_iqtisodiyot_gold_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("etl_catalog.gold.highest_failed_qtisodiyot")
```

```python
language_gold_df = spark.sql("""
SELECT 
    Talim_tili AS language,
    COUNT(*) AS grant_count,
    ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER (), 2) AS grant_percent
FROM etl_catalog.silver.students_cleaned
WHERE status = 'Grant'
GROUP BY Talim_tili
""")

language_gold_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("etl_catalog.gold.grant_by_language")
```
